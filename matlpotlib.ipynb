```python
cpoint = np.array([1,2,32,44,88])
```

```python
plt.plot(cpoint,'o')
plt.show()
```

```python
plt.plot(cpoint,'o:r')
plt.show()
```

```python
plt.plot(cpoint,marker = 'o',ms = 15)
plt.show()
```

```python
plt.plot(cpoint,marker = 'o',ms = 15,mec = 'r')
plt.show()
```

```python
plt.plot(cpoint,marker = 'o',ms = 15,mfc = 'r')
plt.show()
```

```python
plt.plot(cpoint,color = 'r')
plt.show()
```

```python
plt.plot(cpoint,color = 'r',linewidth = '12')
plt.show()
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
plt.plot(xpoints,ypoints)

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid()
```
